•
Tạo một ma trận 3x3 chứa các số nguyên ngẫu nhiên từ 1 
đến 20.
•
Tính giá trị trung bình (mean) của toàn bộ ma trận.
•
Tìm giá trị lớn nhất (max) của mỗi hàng (sử dụng `axis=1`).

```python
import numpy as np
np.random.seed(36)
matrix_3x3 = np.random.randint(1, 36, size=(3, 3))
print("Ma trận 3x3 ngẫu nhiên:")
print(matrix_3x3)
trung_binh = np.mean(matrix_3x3)
print("Giá trị trung bình của ma trận:", trung_binh)
max_hang = np.max(matrix_3x3, axis=1)
print("Giá trị lớn nhất của mỗi hàng:", max_hang)
print(max_hang)
```

Bài tập 4: Đại số Tuyến tính
Sử dụng ma trận `A` từ Bài tập 3:
Tính:
•a) Tích ma trận: `A @ B`
• b) Tích theo từng phần tử: `A * B`

```python
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print (" ma trận a là  : ",a)
print (" ma trận b là  : ",b)

tich = a @ b
print("Tích ma trận a @ b là :", tich)

tich_phantu = a * b
print("Tích theo từng phần tử a * b là :", tich_phantu)
```

Bài tập 4: Đại số Tuyến tính
Sử dụng ma trận `A` từ Bài tập 3:
•a) Ma trận chuyển vị của A (`A.T`).
• b) Ma trận nghịch đảo của A (`np.linalg.inv(A)`)

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])

print("Ma trận a gốc:", a)

a_T = a.T
print("Ma trận chuyển vị :", a_T)

try:
    a_inv = np.linalg.inv(a)
    print("Ma trận nghịch đảo :", a_inv)
    
    test = a @ a_inv
    print("Kiểm tra:", np.round(test))
except np.linalg.LinAlgError:
    print("Ma trận không khả nghịch")
```
